```python
totalComuna = df.loc[(df.comuna == 14)]
houses = df.loc[(df.comuna == 14) & (df.property_type == 'house')]
stores = df.loc[(df.comuna == 14) & (df.property_type == 'store')]
apartments = df.loc[(df.comuna == 14) & (df.property_type == 'apartment')]
phs = df.loc[(df.comuna == 14) & (df.property_type == 'PH')]
```

```python
df['houses'] = (len(houses) * 100) / len(totalComuna)
df['stores'] = (len(stores) * 100) / len(totalComuna)
df['apartments'] = (len(apartments) * 100) / len(totalComuna)
df['phs'] = (len(phs) * 100) / len(totalComuna)
```
